```python
import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
corpus="Hi my name is Yash and today I am learning NLP. And I am typing anything just to make a paragraph but anything will do."
```

```python
#tokenization is the process of breaking down a text into smaller pieces called tokens. These tokens can be words, phrases, or even sentences. Tokenization is an essential step in natural language processing (NLP) as it helps in understanding the structure and meaning of the text.
## Sentence-->paragraphs
from nltk.tokenize import sent_tokenize
sentences=sent_tokenize(corpus)
print(sentences)
```

```python
for sentence in sentences:
    print(sentence)
```

```python
## Tokenization 
## Paragraph-->words
## sentence--->words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in sentences:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
